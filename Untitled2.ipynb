# ------------------------ PART A ----------------------------
# Data Loading & Inspection
# In this section, we load the dataset and perform basic inspection
# to understand its structure, datatypes, and presence of missing values

```python
from google.colab import files
uploaded = files.upload()

import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv('uber_stock_data.csv')

# Display first 5 rows
print("First 5 rows of dataset:\n", df.head())

# Display last 3 rows
print("\nLast 3 rows of dataset:\n", df.tail(3))

# Check data types
print("\nData Types:\n", df.dtypes)

# Check missing values
print("\nMissing Values:\n", df.isna().sum())
```

# ------------------------ PART B ----------------------------
# Data Cleaning
# Cleaning involves handling missing values and ensuring date formats
# are correctly parsed to allow time-based analysis.

```python
df.dropna(inplace=True)

# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])
# Check again after cleaning
print("\nMissing Values After Cleaning:\n", df.isna().sum())
print("\nData Types After Date Conversion:\n", df.dtypes)
```

# ------------------------ PART C ----------------------------
# Basic Analysis (NumPy & Pandas)
# We compute descriptive statistics and use filters, sorting,
# and aggregation to understand trends in the stock data.

```python
print("\nMean Closing Price:", df['Close'].mean())
print("Median Closing Price:", df['Close'].median())
print("Standard Deviation of Close:", df['Close'].std())

# Summary stats
print("\nDescriptive Statistics:\n", df.describe())

# Filter where Close > 60
filtered_df = df[df['Close'] > 60]
print("\nRows where Close > 60:\n", filtered_df.head())

# Sort by Volume
sorted_df = df.sort_values('Volume', ascending=False)
print("\nTop 5 rows sorted by Volume:\n", sorted_df.head())
```

# ------------------------ PART D ----------------------------
# Advanced Operations
# Adding new features, grouping by time-based features,
# and merging with another dataset to simulate real-world joins.

```python
df['Daily Change'] = df['Close'] - df['Open']
df['Month'] = df['Date'].dt.month
avg_close_by_month = df.groupby('Month')['Close'].mean()
print("\nAverage Close by Month:\n", avg_close_by_month)

# Dummy second DataFrame for merge
df2 = df[['Date', 'Volume']].copy()
df2.rename(columns={'Volume': 'Traded Volume'}, inplace=True)
merged_df = pd.merge(df, df2, on='Date')
print("\nMerged Data Sample:\n", merged_df.head())
```

# ------------------------ PART E ----------------------------
# Pivot Tables
# Pivot tables allow multi-dimensional summarization and comparison
# of key statistics like average price or total volume.

```python
pivot1 = pd.pivot_table(df, index='Month', values='Volume', aggfunc='sum')
print("\nPivot Table 1 - Total Volume by Month:\n", pivot1)

# Add year for multi-index pivot
df['Year'] = df['Date'].dt.year
pivot2 = pd.pivot_table(df, index='Month', columns='Year', values='High', aggfunc='mean')
print("\nPivot Table 2 - Avg High Price by Month & Year:\n", pivot2)

grouped_compare = df.groupby(['Month', 'Year'])['High'].mean().unstack()
print("\nGroupBy Comparison:\n", grouped_compare)
```

# ------------------------ PART F ----------------------------
# Data Visualization
# Visual representations help spot patterns, outliers,
# and trends over time or across categories.

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')

# Histogram of Daily Change
plt.figure(figsize=(8, 4))
plt.hist(df['Daily Change'], bins=30, color='skyblue')
plt.title('Distribution of Daily Price Change')
plt.xlabel('Daily Change ($)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# Lineplot of Closing Prices
plt.figure(figsize=(10, 4))
sns.lineplot(x='Date', y='Close', data=df)
plt.title('Closing Price Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.tight_layout()
plt.show()

# Volume vs Close Scatter
plt.figure(figsize=(8, 4))
sns.scatterplot(x='Volume', y='Close', data=df)
plt.title('Volume vs Close Price')
plt.tight_layout()
plt.show()

# Boxplot of Daily Change by Month
plt.figure(figsize=(10, 5))
sns.boxplot(x='Month', y='Daily Change', data=df)
plt.title('Daily Price Change by Month')
plt.tight_layout()
plt.show()
```

# ------------------------ PART G ----------------------------
# Advanced NumPy Operations
# Using NumPy to perform statistical normalization and logical filtering
# for advanced analytics and pattern recognition.

```python
close_mean = np.mean(df['Close'])
close_std = np.std(df['Close'])
df['Close_zscore'] = (df['Close'] - close_mean) / close_std

close_prices = df['Close'].values
mid_range_days = df[np.logical_and(close_prices >= 50, close_prices <= 60)]
print("\nNumber of days with Close between $50 and $60:", len(mid_range_days))
```
